```python
new_spam = []

for i in range(0,len(train_spam)):
    a = train_spam[i].split()
    for b in range(0,len(a)):
        new_spam.append(a[b])

new_ham = []
        
for i in range(0,len(train_ham)):
    a = train_ham[i].split()
    for b in range(0,len(a)):
        new_ham.append(a[b])
        
print(len(new_spam),len(new_ham))
```

No código acima separamos usamos o metódo 'split' para separar as mensagens em palavras para facilitar a classificação pelo classificador.

```python
dict_ham = {}

dict_spam = {}


for word in new_ham:
    if word not in dict_ham:
        dict_ham[word] = [1]
        
    else:
        uhul = dict_ham[word][0]
        dict_ham[word] = [uhul+1]
        
for word in new_spam:
    if word not in dict_spam:
        dict_spam[word] = [1]
        
    else:
        uhul = dict_spam[word][0]
        dict_spam[word] = [uhul+1]
    
```

Contamos a quantidade de vezes que cada palavra aparecia nas mensagens de 'ham' e 'spam' e pudemos classificar quais palavras apareciam mais nas mensagens de 'spam' e quais palavras mais apareciam nas mensagens de 'ham'. As palavra que aparecem o mesmo número de vezes em mensagens 'ham' e 'spam' é declarada 'neutra'.

```python
vezes_ham_spam = {}

for word in dict_spam:
    
    if word in dict_ham:
        
        vezes_spam = dict_spam[word][0]
        vezes_ham = dict_ham[word][0]
        vezes_ham_e_spam = dict_spam[word][0]+dict_ham[word][0]
        
        prob_spam = (vezes_spam/vezes_ham_e_spam)
        prob_ham = (vezes_ham/vezes_ham_e_spam)
        
        if prob_ham > prob_spam:
            vezes_ham_spam[word] = [dict_spam[word][0],dict_ham[word][0],"HAM"]
        if prob_ham < prob_spam:
            vezes_ham_spam[word] = [dict_spam[word][0],dict_ham[word][0],"SPAM"]
        if prob_ham == prob_spam:
            vezes_ham_spam[word] = [dict_spam[word][0],dict_ham[word][0],"NEUTRAL"]
        
vezes_ham_spam  
```
